```python
import numpy as np
import pandas as pd
import os
import random

import warnings
warnings.filterwarnings('ignore')

import prophet
from prophet.diagnostics import cross_validation, performance_metrics

import matplotlib.pyplot as plt

import statsmodels.api as sm

train = pd.read_csv("./data/train.csv")


```

```python
train[''] = pd.to_datetime(train[''])
plt.figure(figsize=(16,8))

plt.plot(train[''], train[''])
plt.show()
```

```python
df_train = train[['','']]
df_train.rename(columns={'': 'ds', '': 'y'}, inplace=True)
```

```python
import itertools

param_grid = {
    'changepoint_prior_scale' : [0.01, 0.05, 0.1, 0.5],
    'seasonality_prior_scale': [0.1, 1.0, 10.0, 15.0],
    'seasonality_mode': ['additive', 'multiplicative']
}

all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]
maes = []

# initial:  , horizon:  , period: cutoff 
# 3  15   1 
for params in all_params:
    m = prophet.Prophet(**params, yearly_seasonality=True).fit(df_train)  # Fit model with given params
    df_cv = cross_validation(m, initial='15Y', period='3Y',horizon = '1Y', parallel="processes")
    df_p = performance_metrics(df_cv, rolling_window=1)
    maes.append(df_p['mae'].values[0])
    
tuning_results = pd.DataFrame(all_params)
tuning_results['mae'] = maes
print(tuning_results)
```

```python
best_params = all_params[np.argmin(maes)]
print(best_params)
```
